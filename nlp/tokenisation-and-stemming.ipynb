```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

text = "the cat sat on the mat the cat sat"
tokens = word_tokenize(text)

# Generate bigrams and count frequency
bigrams = list(ngrams(tokens, 2))
bigram_counts = Counter(bigrams)

# Load into a DataFrame
df = pd.DataFrame(bigram_counts.items(), columns=["bigram", "count"])
df = df.sort_values("count", ascending=False)

print(df)
#          bigram  count
# 0  (the, cat)       2
# 1  (cat, sat)       2
# ...

# Visualize with matplotlib
df.plot(kind="bar", x="bigram", y="count", legend=False)
plt.title("Bigram Frequency")
plt.xlabel("Bigram")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
