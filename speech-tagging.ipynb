```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
emma_ja = "Emma Woodhouse, handsome, clever, and rich, with a comfortable home and happy disposition, seemed to unite some of the best blessings of existence; and had lived nearly twenty-one years in the world with very little to distress or vex her."
print(emma_ja)
```

```python
spacy_doc = nlp(emma_ja)
```

```python
pos_df = pd.DataFrame(columns=['token', 'pos_tag'])
```

```python
for token in spacy_doc:
    pos_df = pd.concat([pos_df, pd.DataFrame.from_records([{'token': token.text, 'pos_tag': token.pos_}])], ignore_index=True)
```

```python
pos_df.head(15)
```

```python
pos_df_counts = pos_df.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
pos_df_counts.head(10)
```

```python
pos_def_poscounts = pos_df_counts.groupby(['pos_tag'])['token'].count().sort_values(ascending=False)
```

```python
pos_def_poscounts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == 'NOUN'][:10]
nouns
```
