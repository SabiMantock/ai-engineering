```python
def count(numbers):
    total = 0
    for x in numbers:
        if x < 20:
            total += 1
    return total
```

```python
list_q=[1,3,7,15,23,43,56,98,17]
```

```python
count(list_q)
```

```python
prices = {
    "box_of_spaghetti" : 4,
    "lasagna"  : 5,
    "hamburger" : 2
   }
quantity = {
    "box_of_spaghetti" : 6,
    "lasagna"  : 10,
    "hamburger" : 0
    }

money_spent = 0
```

```python
for i in quantity:
    if prices[i] >= 5:
        money_spent += prices[i] * quantity[i]
    else:
        money_spent = money_spent

print(money_spent)
```
