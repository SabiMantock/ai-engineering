```python
import spacy
from spacy import displacy
from spacy import tokenizer
from IPython.display import HTML, display
import re
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
google_text = "Google was founded on September 4, 1998, by computer scientist Larry Page and Sergey Brin while they were PhD students at Stanford University in California. Together they own about 14% of its publicly listed shares and control 56% of its stockholder voting power through super-voting stock. the company went public via an intial public offering (IPO) in 2004. In 2015, Google was recognized as a wholly owned subsidiary of Alphabet's largest subsidiary and is a holding company for Alpahbet's internet properties and interests. Sundar Puchai was appointed CEO of Google in October 24, 2015, replacing larry Page, who became CEO of Alphabet. On December 3, 2019,  Pichai also became CEO of Alphabet."
print(google_text)
```

```python
spacy_doc = nlp(google_text)
```

```python
for word in spacy_doc.ents:
    print(word.text, word.label_)
```

```python
html =  displacy.render(spacy_doc, style="ent", jupyter=False)
display(HTML(html))
```

```python
google_text_clean = re.sub(r'[^\w\s]', '',google_text).lower()
print(google_text_clean)
```

```python
spacy_doc_clean = nlp(google_text_clean)
```

```python
for word in spacy_doc_clean.ents:
    print(word.text, word.label_)
```

```python
html = displacy.render(spacy_doc_clean, style="ent", jupyter=False)
display(HTML(html))
```
