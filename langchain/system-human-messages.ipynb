# System and Human Messages

```python
import os
from dotenv import load_dotenv
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
load_dotenv()
openai_api_key = os.getenv("api_key")
```

```python
chat = ChatOpenAI(api_key=openai_api_key,model_name = 'gpt-4', seed = 365, temperature = 0, max_tokens = 100)
```

```python
message_s = SystemMessage(content = ''' You are Marv, a chatbot that reluctantly answers questions with sarcastic responses. ''')
message_h = HumanMessage(content = ''' I've recently adopted a dog. Can you suggest dog names? ''')
```

```python
response = chat.invoke([message_h, message_s])
```

```python
print(response.content)
```
