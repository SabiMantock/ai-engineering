# System and Human Messages

```python
import os
from dotenv import load_dotenv
from langchain_openai.chat_models import ChatOpenAI

from langchain_core.prompts import (AIMessagePromptTemplate, HumanMessagePromptTemplate, ChatPromptTemplate, FewShotChatMessagePromptTemplate)
```

```python
load_dotenv()
openai_api_key = os.getenv("api_key")
```

```python
chat = ChatOpenAI(api_key=openai_api_key,model_name = 'gpt-4', seed = 365, temperature = 0, max_tokens = 100)
```

```python
TEMPLATE_H = '''I've recently adopted a {pet}.
Could you suggest some {pet} names?'''

TEMPLATE_AI = '''{response}'''


message_template_h = HumanMessagePromptTemplate.from_template(template= TEMPLATE_H)
message_template_ai = AIMessagePromptTemplate.from_template(template= TEMPLATE_AI)
```

```python
example_template = ChatPromptTemplate.from_messages([message_template_h,message_template_ai])
```

```python
examples = [{'pet': 'dog', 'response': ''' Oh, absolutely! Because I'm sure your dog is just waiting to hear what a chatbot thinks it should be named. But hey, who am I to judge? How about "Bark Twain" (if its a literary hound)? '''}, {'pet': 'cat', 'response': ''' Oh, absolutely! Because I'm sure your cat is just waiting to hear what a chatbot thinks it should be named. But hey, who am I to judge? How about "Sir Meowsalot"? '''}]
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(examples = examples, example_prompt = example_template, input_variables = ['pet'])
```

```python
chat_template = ChatPromptTemplate.from_messages([few_shot_prompt, message_template_h])
```

```python
chat_value = chat_template.invoke({'pet':'rabbit'})
```

```python
chat_value
```

```python
for i in chat_value.messages:
    print(f'{i.type}: {i.content}')
```

```python
response = chat.invoke(chat_value)
```
