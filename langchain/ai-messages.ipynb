# AI Messages

```python
import os
from dotenv import load_dotenv
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage
```

```python
load_dotenv()
openai_api_key = os.getenv("api_key")
```

```python
chat = ChatOpenAI(api_key=openai_api_key,model_name = 'gpt-4', seed = 365, temperature = 0, max_tokens = 100)
```

```python
message_h_dog = HumanMessage(content = ''' I've recently adopted a dog. Can you suggest dog names? ''')
message_ai_dog = AIMessage(content = ''' Oh, absolutely! Because I'm sure your dog is just waiting to hear what a chatbot thinks it should be named. But hey, who am I to judge? How about "Bark Twain" (if its a literary hound)? ''')
message_h_cat = HumanMessage(content = ''' I've recently adopted a cat. Can you suggest cat names? ''')
message_ai_cat = AIMessage(content = ''' Oh, absolutely! Because I'm sure your cat is just waiting to hear what a chatbot thinks it should be named. But hey, who am I to judge? How about "Sir Meowsalot"?  ''')

message_h_fish = HumanMessage(content = ''' I've recently adopted a fish. Can you suggest fish names? ''')

```

```python
response = chat.invoke([message_h_dog, message_ai_dog, message_h_cat, message_ai_cat, message_h_fish])
```

```python
print(response.content)
```
