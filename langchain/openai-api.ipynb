```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
openai_api_key = os.getenv("api_key")
```

```python
client = OpenAI(api_key=openai_api_key)
```

```python
completion = client.chat.completions.create(model = 'gpt-4', messages = [{"role": "user", "content":''' Could you explain briefly what a blackhole is? '''}], max_tokens=250, temperature=0, seed=365, stream=True)
```

```python
completion
```

```python
for i in completion:
    print(i.choices[0].delta.content, end = "")
```

```python
print(completion.choices[0].message.content)
```
